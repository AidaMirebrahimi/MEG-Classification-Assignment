```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
Test_NR = pd.read_csv('/content/drive/My Drive/Test_All_NR.csv').to_numpy()
Train_NR = pd.read_csv('/content/drive/My Drive/Train_All_NR.csv').to_numpy()
```

```
plt.figure(figsize=(30,10))
plt.plot(Test_NR, label = 'Test')
plt.plot(Train_NR, label = 'Train')
plt.legend()
```

```
max_train = np.max(Train_NR)
min_train = np.min(Train_NR)
avg_train = np.mean(Train_NR)
print('Average Train Accuracy =', avg_train)
print('Min Train Accuracy =', min_train)
print('Max Train Accuracy =', max_train)
```

```
max_test = np.max(Test_NR)
min_test = np.min(Test_NR)
avg_test = np.mean(Test_NR)
print('Average Test Accuracy =', avg_test)
print('Min Test Accuracy =', min_test)
print('Max Test Accuracy =', max_test)
```
